### Bolean Indekslash

```python
## Ismlar
names = np.array(['Hasan', 'Husan', 'Bahodir', 'Bilolxon', 'Bositxon', 'Abdulhay', 'Dovudxon'])
names
```

```python
## data

data = np.random.randn(7, 4)
data
```

```python
names == 'Bahodir'
```

```python
data[names == 'Bahodir']  ## Bahodirga tegishli barcha ma'lumotlarni ajratib olish
```

```python
data[names == 'Bositxon'] ## Bositxonga tegishli barcha ma'lumotlarnin ajratib olish
```

```python
data[names=='Bahodir', 2:] # Bahodirga tegishli oxirgi ikkita ustundagi ma'lumotlarni ajratib olish
```

```python
data[names !='Bahodir']  ## (!=) Bahodirdan tashqari qolgan barcha insonlarni ma'lumotlarini chop qilish
```

```python
data[~(names =='Bahodir')] ## (~) bu ham yuqoridagi cell bilan bir xil natijani chiqaradi
```

```python
mask = (names == 'Bahodir') | (names == 'Bositxon') ## "|" or (yoki) -- bu belgi belgisini ifodalaydi
```

```python
mask
```

```python
data[mask]  ## Bahodir va Bositxonlarga tegishli ma'lumotlarni ajratib olish
```
