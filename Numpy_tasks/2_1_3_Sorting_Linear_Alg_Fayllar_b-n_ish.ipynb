**Sorting** tartiblash

**sort** funksiyasi

```python
## 1 -usul
sorted_array = np.sort(arr)
sorted_array
```

```python
## 2 -usul
arr.sort()
arr
```

```python
## axis = 1 >> Ustun bo'yicha sortlash

col_srt = np.sort(arr2d, axis=1)
col_srt
```

**in1d** usuli

```python
#data

arr1 = np.array([6, 0, 0, 3, 2, 5, 6])
arr2 = np.array([0, 2, 3])
```

```python
np.in1d(arr1, arr2) ## arr2 dagi elementlarni arr1 da borligini tekshiradi
```

**setdiff1d** usuli

```python
names1 = np.array(['Jasur', 'Abdurahmon', 'Hasan', 'Muhammad', 'Sarvar', 'Temur'])
names2 = np.array(['Sarvar', 'Abdurahmon', 'Hasan', 'Temur', 'Sarvar', 'Temur'])
```

```python
np.setdiff1d(names1, names2)
```

```python
np.intersect1d(names1, names2)  # bu intersect1d esa aksincha ikklasida borlarini chiqaradi
```
