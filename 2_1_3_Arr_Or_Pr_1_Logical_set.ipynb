### Shart operatori sifatida `where` dan foydalanish

### Tassavur qiling bizda ikkita ma'lumotlar (massivlar) mavjud bular: `xarr` va `yarr`. Agarda shart `True`ni qanoatlantirsa `xarr` elementini qabul qilgan aksincha esa `yarr` elementini qabul qilgan yangi massiv hosil qilinsin.

```python
# Massivlar yaratib olamiz
xarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5, 2.6]) # xarr
yarr = np.array([4.1, 4.2, 4.3, 4.4, 4.5, 4.6]) # yarr
```

```python
# Shart operatirini ishlatish
condition = np.array([True, False, True, False, True, True])
```

```python
natija = np.where(condition, xarr, yarr) ## where shart operati : agar result == True b/sa : xarr, aks holda yarrr
natija
```

```python
array = np.random.randn(6, 6) ## 6, 6 taxminiy qiymat yaratish
array
```

```python
## where
natija = np.where(array < 0, 6, -6)
natija
```
